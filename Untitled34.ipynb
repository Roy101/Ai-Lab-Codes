```python

import pandas as pd
import numpy as np
import sklearn
 
 
yumm_mush = pd.read_csv('/content/sample_data/mushroom-edibility-classification-dataset.csv')
yumm_mush

yumm_mush.isnull().sum()

from sklearn.impute import SimpleImputer

impute = SimpleImputer(missing_values=np.nan, strategy='mean')

impute.fit(yumm_mush[['cap-shape']])
yumm_mush['cap-shape'] = impute.transform(yumm_mush[['cap-shape']])
impute.fit(yumm_mush[['cap-color']])

yumm_mush['cap-color'] = impute.transform(yumm_mush[['cap-color']])
yumm_mush.isnull().sum()

yumm_mush.info()

yumm_mush['class'].unique()

yumm_mush['bruises'].unique()

from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

yumm_mush['class'] = enc.fit_transform(yumm_mush['class'])


print(yumm_mush[['class']].head())
enc = LabelEncoder()


yumm_mush['bruises'] = enc.fit_transform(yumm_mush['bruises'])

print(yumm_mush[['bruises']].head(10))

yumm_mush=yumm_mush.drop(['Unnamed: 0','veil-type','veil-color','ring-number'], axis=1)
yumm_mush

from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(yumm_mush, yumm_mush['class'] ,test_size=.2,random_state=50)
print(X_train.shape)
print(X_test.shape)

from sklearn.svm import SVC
svc=SVC(kernel='linear')
svc.fit(X_train,y_train)
svcScoreBeforePCA=svc.score(X_test,y_test)

print("Test accuracy SVC {:.2f}".format(svc.score(X_train,y_train)))
print("Test accuracy SVC {:.2f}".format(svc.score(X_test,y_test)))

predictions = svc.predict(X_test)
print(predictions)

from sklearn.metrics import confusion_matrix
matrix_no1=confusion_matrix(predictions, y_test)
print(matrix_no1)

from seaborn import heatmap
heatmap(matrix_no1 , cmap="Paired_r", xticklabels=['L0' ,'L1'], yticklabels=['L0' ,'L1'], annot=True)

from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=30)
forest.fit(X_train,y_train)
RforestBeforePCA=forest.score(X_test,y_test)
print("Test accuracy RandomForest {:.2f}".format(forest.score(X_train,y_train)))
print("Test accuracy RandomForest {:.2f}".format(forest.score(X_test,y_test)))

matrix_no2=confusion_matrix(forest.predict(X_test),y_test)
print(matrix_no2)

from seaborn import heatmap
heatmap(matrix_no2 , cmap="Paired_r", xticklabels=['L0' ,'L1'], yticklabels=['L0' ,'L1'], annot=True)

from sklearn.neural_network import MLPClassifier
neuralNet = MLPClassifier(hidden_layer_sizes=(7),activation='relu',max_iter=1000000)
neuralNet.fit(X_train,y_train)
neuralNetBeforePCA=neuralNet.score(X_test,y_test)
print("Train accuracy Neural {:.2f}".format(forest.score(X_train,y_train)))
print("Test accuracy Neural {:.2f}".format(forest.score(X_test,y_test)))

matrix_no3=confusion_matrix(neuralNet.predict(X_test),y_test)
print(matrix_no3)

heatmap(matrix_no3 , cmap="Paired_r", xticklabels=['L0' ,'L1'], yticklabels=['L0' ,'L1'], annot=True)

yumm_mush.keys

from sklearn.decomposition import PCA
pca=PCA(n_components=8)
PrincipalComponents=pca.fit_transform(yumm_mush)

print(PrincipalComponents)

pca.explained_variance_ratio_

sum(pca.explained_variance_ratio_)

principalDataframe = pd.DataFrame(data=PrincipalComponents)
principalDataframe

mainDataFrame=pd.concat([principalDataframe, yumm_mush[["class"]]], axis=1)
mainDataFrame

x1_train,x1_test,y1_train,y1_test= train_test_split(principalDataframe,yumm_mush['class'],test_size=0.2, random_state=50)
print(x1_train.shape)
print(x1_test.shape)

svc.fit(x1_train,y1_train)
svcAfterPCA=svc.score(x1_test,y1_test)
print("Train accuracy SVC {:.2f}".format(svc.score(x1_train,y1_train)))
print("Test accuracy SVC {:.2f}".format(svc.score(x1_test,y1_test)))

neuralNet.fit(x1_train,y1_train)
neuralNetAfterPCA=neuralNet.score(x1_test,y1_test)
print("Train accuracy neural net {:.2f}".format(neuralNet.score(x1_train,y1_train)))
print("Test accuracy neural net {:.2f}".format(neuralNet.score(x1_test,y1_test)))

forest.fit(x1_train,y1_train)
RforestAfterPCA=forest.score(x1_test,y1_test)
print("Train accuracy forest {:.2f}".format(forest.score(x1_train,y1_train)))
print("Test accuracy forest {:.2f}".format(forest.score(x1_test,y1_test)))

svcMatrix=confusion_matrix(svc.predict(x1_test),y1_test)
print("SVC Matrix:\n",svcMatrix)
forestMatrix=confusion_matrix(forest.predict(x1_test),y1_test)
print("Random Forest:\n",forestMatrix)
neuralMatrix=confusion_matrix(neuralNet.predict(x1_test),y1_test)
print("Neural Network Matrix:\n",neuralMatrix)

x_axis = ["SVC","SVC(PAC)","Neural Net", "Neural Net(PAC)","RFC","RFC(PAC)"]
y_axis = [svcScoreBeforePCA*100,svcAfterPCA*100,neuralNetBeforePCA*100,neuralNetAfterPCA*100,RforestBeforePCA*100,RforestAfterPCA*100]
frame= pd.DataFrame({'Classifier':x_axis,'Accuracy': y_axis})
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20,10))
plt.title("Comparison of test cases With and Without PAC")

ax = sns.barplot(x="Classifier", y="Accuracy", data=frame, palette="plasma")

```
